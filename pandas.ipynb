Data Cleaning

```python
print(df.isnull().sum()) #count of null values in each column
```

- data handling and data deletion two ways to handle the null valus in a dataset
- use data handling if column is important
- use data deletion if column is not important

- handling : fillna
- whole column delete : drop
- null columns delete : dropna

```python
df['age'].fillna(df['age'].mean(),inplace=True) #fill null values in 'age' with mean
```
