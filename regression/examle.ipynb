```python
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.impute import SimpleImputer

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.compose import ColumnTransformer

import seaborn as sns
```

```python
temperature = [[20],[22],[24],[26],[28],[30],[32]]
sales =[10,15,25,40,60,90,130]
```

```python
X = temperature
y = sales
```

```python
X 
```

```python
y
```

```python
# train the dataset
lr = LinearRegression()
lr.fit(X, y)
```

```python
# train polynomial regression model on the whole dataset
pr = PolynomialFeatures(degree=5)
X_poly = pr.fit_transform(X)
lr2 = LinearRegression()
lr2.fit(X_poly, y)
```

```python
# predict
y_pred_poly = lr2.predict(X_poly)
print(y_pred_poly)

y_pred_lr = lr.predict(X)
print(y_pred_poly)
```

```python
#visualize the linear regression 
plt.scatter(X, y, color='red')
plt.plot(X, y_pred_lr, color='blue')
plt.title('Linear Regression')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.legend(['X/y_pred_lr', 'X/y'], title = "temperature/Sales",loc='best',facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, y_pred_poly, color='blue')
plt.title('Polynomial Regression')
plt.xlabel('temperature')
plt.ylabel('Sales')
plt.legend(['X/y_pred_lr', 'X/y'], title = "temperature/sales",loc='best',facecolor='white')
plt.box(False)
plt.show()
```

```python
print(f'Linear reg result: {lr.predict([[6.5]])}')
print(f'Polynomial reg results: {lr2.predict(pr.fit_transform([[6.5]]))}')
```
