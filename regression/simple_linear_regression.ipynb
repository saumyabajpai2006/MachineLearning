```python
import pandas as pd #pandas is a library for data manipulation and analysis
import matplotlib.pyplot as plt #matplotlib is a library for creating static, animated, and interactive visualizations in Python
```

```python
df = pd.read_csv('Salary_Data.csv')
df
```

```python
df.isnull().sum()  # Check for missing values in the dataset
```

LINEAR REGRESSION
- Relationship between the dependent and independent variable
- helps in predict and forecast outcomes
- involves only one independent variable

```python
X= df.iloc[:,:-1].values #ilocis used to select rows and columns by index
print(X)
```

```python
y = df.iloc[:,1].values
print(y)
```

MODEL TRAINING

```python
from sklearn.model_selection import train_test_split 

#train_test_split is used to split the dataset into training and testing sets 
#sklearn is a library for machine learning in Python
#model_selection is a module in sklearn that contains functions for splitting datasets into training and testing sets
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
# test size ka matlab 20% of data   
# random state is used to get the same result every time we run the code
```

TRAINING THE MODEL

```python
from sklearn.linear_model import LinearRegression


regressor = LinearRegression()
regressor.fit(X_train, y_train) #regressor.fit is used to train the model on the training data
```

PREDICTING THE RESULT

```python
y_pred = regressor.predict(X_test)
```

```python
print(regressor.predict([[12]]))        # double brackets for 2D array input
```

```python
print(regressor.predict([[2]]))
```

VISUALISING THE TRAINING RESULT

```python
plt.scatter(X_train, y_train,color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_test, y_test,color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Testing set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from joblib import dump,  load #dump is used to save the model and load is used to load the model
with open('models/simple_linear_regression.pkl', 'wb') as f:
    dump(regressor, f)  # Save the model to a file
    print("Model dumped successfully.")
```
