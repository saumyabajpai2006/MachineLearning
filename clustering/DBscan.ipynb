```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Mall_customers.csv')
df.head()  # Load your dataset
```

```python
X = df.iloc[:, [3, 4]].values  # Assuming the last two columns are the features for clustering
```

```python
X = StandardScaler().fit_transform(X)  # Standardizing the features
```

- eps- Neighbourhood radius: How close two points must be in the same cluster
- min_samples- Minimum points in a cluster : The minimum number of points needed to form a cluster

```python
dbscan = DBSCAN (eps=0.5, min_samples=5)        #dbscan clustering #eps is the maximum distance between two samples for one to be considered as in the neighborhood of the other.
labels = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis',s=50)    #cmap='viridis' is used to color the clusters
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
#finding outliers
outliers = X[labels == -1]
print(f"Total outliers detected:, {len(outliers)}")
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis',s=50)
plt.scatter(outliers[:, 0], outliers[:, 1], color='red', s=100, label='Outliers',marker='x')
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
